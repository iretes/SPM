# Performance analysis

Loading sequential times:

```python
seq_data = pd.read_csv("./results/word_count_log_seq.csv", header=None)
seq_data.columns = ["extraworkXline", "computing_time", "sorting_time"]
seq_data["T"] = 1
seq_data
```

Computing median sequential time:

```python
seq_times = {
    0: seq_data.groupby("extraworkXline")['computing_time'].median().loc[0],
    1000: seq_data.groupby("extraworkXline")['computing_time'].median().loc[1000],
    10000: seq_data.groupby("extraworkXline")['computing_time'].median().loc[10000],
}
```

Defining a function to read parallel execution times:

```python
def read_data(path):
    df =  pd.read_csv(path, header=None)
    df.columns = ["Lw", "Rw", "ondemand", "extraworkXline", "map_time", "reduce_time", "sorting_time"]
    df["map_time"] = df["map_time"] / 1000
    df["reduce_time"] = df["reduce_time"] / 1000
    df["sorting_time"] = df["sorting_time"] / 1000
    df["computing_time"] = df["map_time"] + df["reduce_time"]
    return df
```

Reading parallel execution times when using more Rw:

```python
par_data_morerw = read_data("./results/word_count_log_morerw.csv")
par_data_morerw["T"] = par_data_morerw["Lw"] + par_data_morerw["Rw"]
# reordering columns
par_data_morerw = par_data_morerw[["Lw", "Rw", "T", "ondemand", "extraworkXline", "map_time", "reduce_time", "computing_time", "sorting_time"]]
par_data_morerw
```

```python
par_data_morerw[par_data_morerw['extraworkXline']==1000].boxplot(column="computing_time", by=["Lw", "Rw", "ondemand"], figsize=(20, 10), rot=90)
plt.ylabel("Computing time [s]");
plt.suptitle("Computing time for extraworkXline=1000")
plt.title("")
plt.tight_layout()
plt.savefig(f"img/boxplot_times_1000.png")
```

```python
par_data_morerw[par_data_morerw['extraworkXline']==10000].boxplot(column="computing_time", by=["Lw", "Rw", "ondemand"], figsize=(20, 10), rot=90)
plt.ylabel("Computing time [s]");
plt.suptitle("Computing time for extraworkXline=10000")
plt.title("")
plt.tight_layout()
plt.savefig(f"img/boxplot_times_10000.png")
```

Computing median computation time:

```python
# computing median time
par_data_morerw_grouped = par_data_morerw.groupby(["Lw", "Rw", "T", "ondemand", "extraworkXline"]).agg(["median"])
# flattening index
par_data_morerw_grouped.columns = [f"{a}_{b}" for a, b in par_data_morerw_grouped.columns.to_list()]
# removing 'Rw' from index
par_data_morerw_grouped = par_data_morerw_grouped.reset_index().set_index(["Lw", "T", "ondemand", "extraworkXline"])
par_data_morerw_grouped
```

Visualizing reduce time:

```python
par_data_morerw.groupby(["Rw"]).agg(["mean", "std"])["reduce_time"].plot(y="mean", yerr="std", marker="o")
rw = par_data_morerw["Rw"].unique()
rw.sort()
plt.xticks(rw)
plt.gca().get_legend().remove()
plt.grid()
plt.xlabel("# Right Workers")
plt.ylabel("Time [s]")
plt.title(f"Reduce time");
plt.tight_layout()
plt.savefig(f"img/reduce_time.png")
```

Reading and computing the median execution time of the parallel implementation when using equal sized files:

```python
par_balaced_data = read_data('./results/word_count_log_balanced.csv')
par_balanced_time_lw1 = par_balaced_data[par_balaced_data['Lw']==1]['computing_time'].median()
par_balanced_time_lw4 = par_balaced_data[par_balaced_data['Lw']==4]['computing_time'].median()
par_balanced_time_lw14 = par_balaced_data[par_balaced_data['Lw']==14]['computing_time'].median()
par_balanced_time_lw1, par_balanced_time_lw4, par_balanced_time_lw14
```

Reading and computing the median execution time of the sequential implementation when using equal sized files:

```python
seq_balanced_data = pd.read_csv('./results/word_count_log_seq_balanced.csv', header=None)
seq_balanced_data.columns = ["extraworkXline", "computing_time", "sorting_time"]
seq_balanced_time = seq_balanced_data["computing_time"].median()
seq_balanced_time
```

Reading and computing the median execution time of the parallel implementation when using more Lw:

```python
par_data_morelw = read_data('./results/word_count_log_morelw.csv')
par_data_morelw["T"] = par_data_morelw["Lw"] + par_data_morelw["Rw"]
par_data_morelw_grouped = par_data_morelw[["T", "extraworkXline", "computing_time"]].groupby(["extraworkXline", "T"]).agg(["median"])
par_data_morelw_grouped.columns = [f"{a}_{b}" for a, b in par_data_morelw_grouped.columns.to_list()]
par_data_morelw_grouped
```

Plotting execution times:

```python
fig, axs = plt.subplots(1, 3, figsize=(20, 5))
# color palette
colors = plt.cm.tab20.colors
# getting number of threads
threads = np.unique(par_data_morerw_grouped.index.get_level_values("T").to_list())
for i, work in enumerate([0, 1000, 10000]): # for each extraworkXline
    color_id = 0
    # plotting times with more Rw
    for lw in par_data_morerw_grouped.index.get_level_values("Lw").unique(): # for each Lw
        for ondemand in [0, 1]: # for each scheduling policy
            par_data_morerw_grouped["computing_time_median"].loc[lw, :, ondemand, work].plot(
                ax=axs[i],
                label=f"#Lw={lw}, ondemand={ondemand}",
                style="o-",
                color=colors[color_id]
            )
            color_id += 1
    # plotting times with more Lw
    par_data_morelw_grouped["computing_time_median"].loc[work,:].plot(
        ax=axs[i],
        label=f"#Lw=ceil(#Threads/3)*2, ondemand={ondemand}",
        style="o-",
        color=colors[color_id]
    )
    axs[i].set_xlabel("#Threads")
    axs[i].set_ylabel("Time [s]")
    axs[i].set_xticks(ticks=threads, labels=threads)
    axs[i].set_title(f"extraworkXline = {work}")
    axs[i].grid()
axs[2].legend(bbox_to_anchor=(1.05, 1), loc='upper left');
plt.tight_layout()
plt.savefig(f"img/times.png")
```

Defining a function to compute speedup, cost and efficiency:

```python
def compute_metrics(df, seq_time):
    speedup = seq_time / df["computing_time_median"]
    speedup.name = "speedup"

    cost = df["computing_time_median"]*df.index.get_level_values("T")
    cost.name = "cost"

    metrics_df = pd.concat([
        df.index.get_level_values("T").to_series(),
        speedup,
        cost,
        ],
        axis=1
    )

    metrics_df["efficiency"] = metrics_df["speedup"] / metrics_df["T"]

    return metrics_df
```

Plotting speedup:

```python
fig, axs = plt.subplots(1, 3, figsize=(20, 5))
# color palette
colors = plt.cm.tab20.colors
for i, work in enumerate([0, 1000, 10000]): # for each extraworkXline
    color_id = 0
    for lw in par_data_morerw_grouped.index.get_level_values("Lw").unique(): # for each Lw
        # plotting speedup with more Rw
        for ondemand in [0, 1]: # for each scheduling policy
            metrics = compute_metrics(par_data_morerw_grouped.loc[lw, :, ondemand, work], seq_times[work])
            metrics.plot(
                x="T", y="speedup",
                label=f"#Lw={lw}, ondemand={ondemand}", style='o-', color=colors[color_id],
                ax=axs[i]
            )
            color_id += 1
    # plotting speedup with more Lw
    morelw_metrics = compute_metrics(par_data_morelw_grouped.loc[work,:], seq_times[work])
    morelw_metrics.plot(
        x="T", y="speedup",
        label=f"#Lw=ceil(#Threads/3)*2, ondemand={ondemand}", style='o-', color=colors[color_id],
        ax=axs[i]
    )
    axs[i].set_xlabel("#Threads")
    axs[i].set_ylabel("Speedup")
    axs[i].set_xticks(ticks=threads, labels=threads)
    axs[i].set_title(f"extraworkXline = {work}")
    axs[i].grid()
    axs[i].get_legend().remove()
# plotting speedup for balanced workload
axs[2].scatter(
    [28], [seq_balanced_time/par_balanced_time_lw1],
    marker='s', edgecolor='white', zorder=10, s=100, color='tab:blue',
    label='#Lw=1, ondemand=0\nbalanced file size'
)
axs[2].scatter(
    [28], [seq_balanced_time/par_balanced_time_lw4],
    marker='s', edgecolor='white', zorder=10, s=100, color='tab:orange',
    label='#Lw=4, ondemand=0\nbalanced file size'
)
axs[2].scatter(
    [28], [seq_balanced_time/par_balanced_time_lw14],
    marker='s', edgecolor='white', zorder=10, s=100, color='tab:green',
    label='#Lw=14, ondemand=0\nbalanced file size'
)
axs[2].legend(bbox_to_anchor=(1.05, 1), loc='upper left');
plt.tight_layout()
plt.savefig(f"img/speedup.png")
```

Plotting execution cost:

```python
fig, axs = plt.subplots(1, 3, figsize=(20, 5))
# color palette
colors = plt.cm.tab20.colors
for i, work in enumerate([0, 1000, 10000]): # for each extraworkXline
    color_id = 0
    for lw in par_data_morerw_grouped.index.get_level_values("Lw").unique(): # for each Lw
        # plotting cost with more Rw
        for ondemand in [0, 1]: # for each scheduling policy
            metrics = compute_metrics(par_data_morerw_grouped.loc[lw, :, ondemand, work], seq_times[work])
            metrics.plot(
                x="T", y="cost",
                label=f"#Lw={lw}, ondemand={ondemand}", style='o-', color=colors[color_id],
                ax=axs[i]
            )
            color_id += 1
    # plotting cost with more Lw
    morelw_metrics = compute_metrics(par_data_morelw_grouped.loc[work,:], seq_times[work])
    morelw_metrics.plot(
        x="T", y="cost",
        label=f"#Lw=ceil(#Threads/3)*2, ondemand={ondemand}", style='o-', color=colors[color_id],
        ax=axs[i]
    )
    axs[i].set_xlabel("#Threads")
    axs[i].set_ylabel("Cost")
    axs[i].set_xticks(ticks=threads, labels=threads)
    axs[i].set_title(f"extraworkXline = {work}")
    axs[i].grid()
    axs[i].get_legend().remove()
axs[2].legend(bbox_to_anchor=(1.05, 1), loc='upper left');
plt.tight_layout()
plt.savefig(f"img/cost.png")
```

Plotting efficiency:

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5), sharey=True)
for i, work in enumerate([0, 1000, 10000]): # for each extraworkXline
    efficiency_df = pd.DataFrame()
    efficiency_strings = []
    # plotting efficiency with more Rw
    for lw in par_data_morerw_grouped.index.get_level_values("Lw").unique(): # for each Lw
        non_ondemand_metrics = compute_metrics(par_data_morerw_grouped.loc[lw, :, 0, work], seq_times[work])
        non_ondemand_metrics.rename(columns={'efficiency': f'efficiency_#Lw={lw}'}, inplace=True)
        efficiency_df = pd.concat(
            [efficiency_df, non_ondemand_metrics[f'efficiency_#Lw={lw}']],
            axis=1
        )
        efficiency_strings.append(f'efficiency_#Lw={lw}')
    efficiency_df.index.name = 'T'
    efficiency_df.reset_index(inplace=True)
    labels = [str[str.find('_')+1:] for str in efficiency_strings]
    efficiency_df.plot.bar(
        x="T", y=efficiency_strings,
        label=labels,
        ax=axs[i]
    )
    axs[i].set_xlabel("#Threads")
    axs[i].set_ylabel("Efficiency")
    axs[i].set_axisbelow(True)
    axs[i].grid()
    axs[i].set_title(f"extraworkXline = {work}, ondemand=0");
plt.tight_layout()
plt.savefig(f"img/efficiency.png")
```
